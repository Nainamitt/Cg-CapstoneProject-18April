```python
import pandas as pd
import random
```

```python

locations = ["Delhi", "Mumbai", "Bangalore", "Chennai", "Pune", "Kolkata"]
claim_types = ["Health", "Vehicle", "Property"]

data = []

```

```python
for i in range(500):
    claim_amount = random.randint(5000, 200000)
    prev_claims = random.randint(0, 5)
    location = random.choice(locations)
    claim_type = random.choice(claim_types)
    
    # Fraud logic (IMPORTANT for ML learning)
    fraud = "Yes" if (claim_amount > 70000 and prev_claims >= 2) else "No"
    
    data.append([
        f"C{i+1:04d}",
        claim_amount,
        claim_type,
        location,
        prev_claims,
        fraud
    ])

df = pd.DataFrame(data, columns=[
    "Customer_ID",
    "Claim_Amount",
    "Claim_Type",
    "Location",
    "Previous_Claims",
    "Fraud"
])

df.to_csv("insurance_claims_large.csv", index=False)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Claim_Type"] = le.fit_transform(df["Claim_Type"])
df["Location"] = le.fit_transform(df["Location"])
df["Fraud"] = le.fit_transform(df["Fraud"])
```
